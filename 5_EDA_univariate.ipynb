```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
import seaborn as sns
```

**1. Categorical Data**

a. Countplot

```python
sns.countplot(x='Survived', data = df)
df['Survived'].value_counts()
#549 people died
```

```python
sns.countplot(x='Pclass', data = df)
```

```python
sns.countplot(x='Embarked', data = df)
```

b. PieChart

```python
df['Survived'].value_counts().plot(kind = 'pie', title='Survived' , autopct = '%.2f')
```

```python
df['Pclass'].value_counts().plot(kind = 'pie', title='PClass' , autopct = '%.2f')
```

**2. Numerical Data**

a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

b. Distplot

```python
sns.distplot(df['Age']) #Probability Density
```

c. Boxplot

```python
sns.boxplot(df['Fare'])
```
